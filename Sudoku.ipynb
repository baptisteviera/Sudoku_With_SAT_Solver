#### Création des contraintes liées aux règles Sudoku

Chaque cellule doit avoir une unique valeur donc faire une combinaison

1 + (9*8)/2 = 37
le 1 correspond à :
1;1=1 or 1;1=2 or 1;1=3 or 1;1=4 or 1;1=5 or 1;1=6 or 1;1=7 or 1;1=8 or 1;1=9

Ensuite le (9*8)/2 correspond à :
(n(n-1))/2
Car au début on a n, puis n-1, puis n-2 .... car il y a l'unicité si une valeur est placé alors on ne peut plus utiliser cette valeur

```python
"""
Version 1
"""

def create_cell_constraints() -> List[List[int]]:
    list_tot = []
    for i in range (1,729,9):
        list_temp = []
        for j in range (i,i+9):
            list_temp.append(j)
        list_tmp_unique = unique(list_temp)
        for k in list_tmp_unique:
            list_tot.append(k)
    return list_tot
```

```python
def exec_gophersat_nb_solution(
    filename: str, cmd: str = "./gophersat-1.1.6", encoding: str = "utf8"
):
    result = subprocess.run(
        [cmd, "--count",filename], capture_output=True, check=True, encoding=encoding
    )
    string = str(result.stdout) # 'c solving sudoku2.cnf\n1\n'
    lines = string.splitlines()
    if len(lines) == 2:
        return "Nombre de modèles unique "+lines[1]
    else:
        return "Nombre de solutions non unique"

```
